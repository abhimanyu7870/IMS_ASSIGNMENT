## Read Data from JSON

```python
import json
```

```python
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

## Add New Item into Inventory

```python
Product_id = str(input("Enter product id:"))
if Product_id  in record:
    print(" Enter New Product ID")
else:
    Product_name = str(input("Enter Product name:"))
    Price = int(input("Enter price:"))
    
    #All weights are in gm:
    Weight = int(input("Enter Weight:"))
    Brand = str(input("Enter the brand name:"))
    Category = str(input("Enter Category:"))
    Quantity = int(input("Enter Quantity:"))
```

```python
 record[Product_id] = {'Product name': Product_name, 'Price': Price, 'Weight': Weight, 'Brand': Brand, 
                      'Category': Category, 'Quantity': Quantity}
```

```python
js = json.dumps(record)
```

```python
fd = open("record.json",'w')
fd.write(js)
fd.close()
```
